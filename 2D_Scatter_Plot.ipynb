```python
import pandas as pd
import plotly.express as px
import numpy as np
import re
import plotly.graph_objects as go
```

```python
df = pd.read_csv('data/02-processed/dataset_3_cleaned.csv')
```

```python
df.head()
```

```python
def extract_year_range(x):
    if "(" in x:
        inside = x.split("(")[1].split(")")[0]
        x = inside.strip()
    if "-" in x:
        values = x.split('-')
        return int(values[1]) - int(values[0])
    return 0
```

```python
df['years_operated'] = df['Years.of.Operation'].apply(extract_year_range)
```

```python
df
```

```python
df['How.Much.They.Raised'] = df['How.Much.They.Raised'].apply(lambda x : x.strip('$'))
```

```python
def converter(x):
    if not isinstance(x, str):
        return np.nan  
    x = x.upper().strip()

    match = re.findall(r"[\d\.]+", x)
    if not match:
        return np.nan
    
    num = float(match[0])

    if "B" in x:
        return num * 1_000_000_000
    elif "M" in x:
        return num * 1_000_000
    else:
        return num
```

```python
df['How.Much.They.Raised'] = df['How.Much.They.Raised'].apply(converter)
```
